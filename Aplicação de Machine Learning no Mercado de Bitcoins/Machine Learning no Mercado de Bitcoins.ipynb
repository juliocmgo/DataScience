## 2) Introdução

Bitcoin é um criptomoeda que surgiu em 2009, e nos últimos anos tem chamado bastante atenção, por ter atingido uma valorização bastante alta. Sua criação é atribuída a Satoshi Nakamoto, quando em 2008, publicou o artigo, *Bitcoin: A Peer-to-Peer Electronic Cash System*<sup>1</sup>, que tornou conhecido o conceito do bitcoin.

O processo de nascimento de uma bitcoin, chamado de “mineração”, nesse processo os mineradores competem entre si para resolver um problema criptográfico. Quando uma solução é encontrada, quem encontrou anuncia aos demais nós na rede e reivindica o prêmio em bitcoins.

Outra maneira de se adquirir bitcoins é através da compra, para isso é necessário ter uma carteira digital, onde as moedas serão armazenadas, e com isso pode-se obter por meio de Bolsas de Bitcoin (*Exchanges*) ou oferecer serviços em troca de bitcoins como pagamento.

Atualmente diversos estabelecimentos já reconhecem a criptomoeda como forma de pagamento. Podemos citar como exemplos, no Japão é possível pagar contas de luz, comprar em sites internacionais como o Ebay, Dell ou Microsoft, adquirir passagens aéreas, fazer recargas de celular e pagar boletos com o serviço da Pague com Bitcoin ou o BitBol, entre outras aplicações que podem ser feitas com a moeda<sup>2</sup>.

# KNN

```python
random_indices = permutation(BitcoinHistoric.index)
#Separando o conjunto de dados entre 70% para treinamento e 30% para teste.
train_df_knn = BitcoinHistoric.iloc[random_indices[0:int(0.7*len(BitcoinHistoric))]]

test_df_knn = BitcoinHistoric.iloc[int(0.7*len(BitcoinHistoric))+1:]

Date_val_knn = test_df_knn['Date'][5:]
Date_val_knn = [pd.to_datetime(d) for d in Date_val_knn]
```

```python
#Usando o grid search para otimizar os hyper parametros (os K vizinhos), para encontrar o valor de K com o menor erro.
#Inicialmente realizando a procure com três caractericas (coluna) do dataset
hyper_params = [x for x in range(1,41)]
mse_values = []

for hp in hyper_params:
    knn = KNeighborsRegressor(n_neighbors=hp, weights='distance',algorithm='auto')

    train_columns = ['Open','High','Low']
    train = train_df_knn[train_columns]

    target_columns = 'Close'
    target = train_df_knn[target_columns]

    knn.fit(train, target)

    predictions = knn.predict(test_df_knn[train_columns])
    mse_values.append(mean_squared_error(test_df_knn['Close'],predictions))

print(np.sqrt(mse_values))    

sns.regplot(x = np.array(hyper_params), y = np.array(np.sqrt(mse_values) ))
plt.show()
```

```python
#Obtendo o valor K com menor erro para três parametros
lowest_mse_3_params = np.min(mse_values)
lowest_mse_3_params_index = mse_values.index(lowest_mse_3_params)  + 1
lowest_mse_3_params_index
```

```python
#Realizando a procure com seis caractericas (coluna) do dataset
hyper_params = [x for x in range(1,41)]
mse_values = []

for hp in hyper_params:
    knn = KNeighborsRegressor(n_neighbors=hp, weights='distance',algorithm='auto')

    train_columns = ['Open', 'High', 'Low', 'Volume (BTC)', 'Volume (Currency)', 'Weighted Price']
    train = train_df_knn[train_columns]

    target_columns = 'Close'
    target = train_df_knn[target_columns]

    knn.fit(train, target)

    predictions = knn.predict(test_df_knn[train_columns])
    mse_values.append(mean_squared_error(test_df_knn['Close'],predictions))

print(np.sqrt(mse_values))    

sns.regplot(x = np.array(hyper_params), y = np.array(np.sqrt(mse_values)) )
plt.show()
```

```python
lowest_mse_6_params = np.min(mse_values)
lowest_mse_6_params_index = mse_values.index(lowest_mse_6_params) + 1 
lowest_mse_6_params_index 
```
